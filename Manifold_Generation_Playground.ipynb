```python
import skdim.datasets as skd
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

%matplotlib widget
```

# Demonstrating skdim.datasets's manifold functions

## hyperSphere

- Note that the dimension parameter is one less than that which it's projected into, as the manifold of a sphere within D=3 is of dimension 2
- Does not have a center or radius parameter

```python
data = skd.hyperSphere(300, 1)
fig = plt.figure()
ax = fig.add_subplot(111)
ax.scatter(data[:,0], data[:,1])
```

```python
data = skd.hyperSphere(300, 2)
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.scatter(data[:,0], data[:,1], data[:,2])
```

## hyperBall
- It's like the sphere, but filled in.

```python
data = skd.hyperBall(300, 2, center=(1,1), radius=1.5)
fig = plt.figure()
ax = fig.add_subplot(111)
ax.scatter(data[:,0], data[:,1])
```

```python
data = skd.hyperBall(300, 3, center=(1,1,1), radius=3)
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.scatter(data[:,0], data[:,1], data[:,2])
```

## hyperTwinPeaks

```python
data = skd.hyperTwinPeaks(300, 1, height=2.5)
fig = plt.figure()
ax = fig.add_subplot(111)
ax.scatter(data[:,0], data[:,1])
```

```python
data = skd.hyperTwinPeaks(600, 2, height=2.5)
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.scatter(data[:,0], data[:,1], data[:,2])
```

## lineDiskBall
- Returns a tuple, one with 3D data, and the other specifying which part of the line, disk, or ball 

```python
data = skd.lineDiskBall(600)
print(len(data))
print(data[0].shape)
print(data[1].shape)
colors = data[1]
data = data[0]
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.scatter(data[:,0], data[:,1], data[:,2], c=colors)
```

## swissRoll3Sph
- A swiss roll with a 3-sphere at the center.
- swissRoll3Sph(num_swiss_pts, num_sphere_pts, a=internal_radius, b=external_radius, nturn=num_turns, h=roll_height)

```python
data = skd.swissRoll3Sph(600, 0, a=3.5, b=4, nturn=3)

data.shape

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.scatter(data[:,0], data[:,1], data[:,2])
```
